```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
data=requests.get("https://www.xe.com/currencyconverter/convert/?Amount=1&From=USD&To=EGP")
soup=BeautifulSoup(data.text,'html.parser')
soup
```

```python
test=soup.find_all('div',attrs={'class':'sc-b89b6e5-1 hhMyNb !m-0 !w-full !bg-transparent !shadow-none'})
test[0]
```

```python
USD=test[0].find_all('div',attrs={'class':'flex justify-center gap-2 font-semibold'})
USDtitle=USD[0].text
EGP=USD[1].text
EGP
number=test[0].find_all('a',attrs={'class':'text-center font-semibold text-blue-500'})
number_=[]
for i in range(len(number)):
    number_.append(number[i].text.strip("USD"))
number_
```

```python
converted=test[0].find_all('div',attrs={'class':'text-center font-semibold text-greyblue-400'})
converted

```

```python
oneusdlink=test[0].find_all('a',attrs={'class':'text-center font-semibold text-blue-500'})
numbersconverted=[]
for i in range(len(oneusdlink)):
    print("scraping",i+1)
    onelink='https://www.xe.com'+oneusdlink[i]['href']
    openlink=requests.get(onelink)
    soup2=BeautifulSoup(openlink.text,'html.parser')
    result=soup2.find_all('p',attrs={'class':'sc-b5d1d4ba-1 bPeOTN'})
    x=result[0].text.strip('Egyptian Pounds').replace(',','').strip()
    numbersconverted.append(x)
    print(numbersconverted)
```

```python
with open('exchangerates.csv','w',encoding='utf8') as fd:
    fd.write('USD,EGP\n')
    USD=test[0].find_all('div',attrs={'class':'flex justify-center gap-2 font-semibold'})
    USDtitle=USD[0].text
    EGP=USD[1].text
    EGP
    number=test[0].find_all('a',attrs={'class':'text-center font-semibold text-blue-500'})
    number_=[]
    for i in range(len(number)):
        number_.append(number[i].text.strip("USD"))
    oneusdlink=test[0].find_all('a',attrs={'class':'text-center font-semibold text-blue-500'})
    numbersconverted=[]
    for i in range(len(oneusdlink)):
        print("scraping",i+1)
        onelink='https://www.xe.com'+oneusdlink[i]['href']
        openlink=requests.get(onelink)
        soup2=BeautifulSoup(openlink.text,'html.parser')
        result=soup2.find_all('p',attrs={'class':'sc-b5d1d4ba-1 bPeOTN'})
        x=result[0].text.strip('Egyptian Pounds').replace(',','').strip()
        numbersconverted.append(x)
        fd.write(f"\"{number_[i]}\",\"{numbersconverted[i]}\"\n")
     
   
```
